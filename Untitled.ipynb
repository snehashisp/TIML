### Reinforcement Learning Structures

This part contains class definitions for the various parts of a Reinforcement Learning Model, namely Actions, States, Policy as well as the Environment and the Agent which interact with them.

```python
import numpy as np
import copy
import matplotlib.pyplot as plt
import math
from mpl_toolkits import mplot3d
#Action Defines an action on a State
class Action():

    def __init__(self, action_name = None):
        self.action_name = action_name
    
    def __hash__(self):
        return hash(self.action_name)
    def __eq__(self, other):
        if other == None:
            return False
        return self.action_name == other.action_name
    def __repr__(self):
        return str(self.action_name)

#Definition of the states of the envirionment
class State():
    
    #states are associated with the transitions and the rewards.
    #note that each state must have a unique state name
    def __init__(self, state_name, reward = None):
        self.state_name = state_name
        self.transitions = {}
        self.reward = reward
        self.state_value = 0
        self._action_reward = {}
    
    #get the next state based on the action done and the transition probabilities
    def nextState(self, action):
        next_states = self.transitions.get(action, {})
        next_state, reward = None, 0
        if next_states != {}:
            next_state = np.random.choice(list(next_states.keys()), 
                                p = list(map(lambda x:x[0],next_states.values())))
            reward = next_states[next_state][1]
        return next_state, reward
    
    #expected reward given an action.
    def getActionReward(self, action):
        reward = self._action_reward.get(action, None)
        if reward is not None:
            return reward
        reward = 0
        for state, prob_reward in self.transitions.get(action, {}).items():
            reward += prob_reward[0] * (state.reward + prob_reward[1])
        self._action_reward[action] = reward
        return reward
    
    def getStateTransitionProb(self, state, action):
        return self.transitions.get(action, {}).get(states, 0)
        
    #add a transition on this state on given action with a certain probability
    def addTransition(self, action, state, prob, trans_reward = 0):
        action_states = self.transitions.setdefault(action, {})
        action_states.setdefault(state, [prob, trans_reward])
            
        
    def __hash__(self):
        return hash(self.state_name)
    def __eq__(self,other):
        if other == None:
            return False
        return self.state_name == other.state_name
    def __repr__(self):
        return str(self.state_name)
    
```

```python
#Policy defines the probability of choosing an action given a state
class Policy():

    def __init__(self):
        self.policy = {}
    
    #update or add the probabilites of taking an action on a state
    def updatePolicy(self, state, action, prob, reset = False):
        if reset:
            self.policy[state] = {}
        state_actions = self.policy.setdefault(state, {})
        state_actions.update({action:prob})
    
    def getActionProb(self, state, action):
        return self.policy.get(state, {}).get(action, 0)
        
    #generate a randon policy given a set of states and actions
    def generateRandomPolicy(self, states, actions, rand_func = np.random.uniform, **rand_func_args):
        for state in states:
            for action in actions:
                self.addPolicy(state, action, rand_func())
    
    #get action based on the probabilites on actions defined on a state
    def getAction(self, state):
        actions = self.policy.get(state, {})
        action = None
        if actions != {}:
            action = np.random.choice(list(actions.keys()), p = list(actions.values()))
        return action
```

```python
"""
Environment class defines an environment on which contains a set of states.
It also defines the discount factor and the cmulative rewards for interacting with the environment using actions.
"""

class Environment():

    def __init__(self, discount = 0.9):
        self.states = set()
        self.current_state = None
        self.discount = discount
        self.discount_multiplier = discount
        self.cumulative_reward = 0

    #add a state to the environment
    def addState(self, state):
        self.states.add(state)
    
    #do action on the current state
    def step(self, action):
        if not self.current_state:
            return
        next_state, reward = self.current_state.nextState(action)
        if next_state == None:
            return
        self.current_state = next_state
        self.cumulative_reward += self.discount_multiplier * (self.current_state.reward + reward)
        self.discount_multiplier = self.discount * self.discount_multiplier
    
    #reset the rewards
    def reset(self, init_state = None):
        self.discount_multiplier = self.discount
        self.cumulative_reward = 0
        for state in self.states:
            state.state_value = 0
            state._action_reward = {}
```

```python
#Agent interacts with the environment 
class Agent():

    def __init__(self, env, policy):
        self.env = env
        self.policy = policy
    
    #get the required action from the policy based on the current state and use action on the state
    def interact(self):
        action = self.policy.getAction(self.env.current_state)
        if action is not None:
            self.env.step(action)
        return self.env.cumulative_reward
```

```python
def simulateAgent(agent, start_state, end_states, showRes = True):
    agent.env.current_state = start_state
    if showRes:
        print("State Transitions")
    while agent.env.current_state not in end_states:
        if showRes:
            print(agent.env.current_state.state_name, end = " ")
        agent.interact()
    if showRes:
        print("\nCumulative Reward")
        print(agent.env.cumulative_reward)
    #agent.env.reset()
    return agent.env.cumulative_reward
```

```python
#Set up the states
gridHeight, gridWidth = 8,8
states = [[State(str((j,i)),0) for i in range(gridWidth)] for j in range(gridHeight)]
states[gridHeight - 1][gridWidth - 1].reward = 1
#Set up the actions
up, down, left, right = Action("U"), Action("D"), Action("L"), Action("R")
#Random Policy that allows all actions with equal probability
random_policy = Policy()
```

All transition on a state and given action is deterministic i.e the agent will always move to the corresponding cell defined by the action if it is possible. The action probability for each state, defined in a policy, to all other possible states is set to be equal i.e. if there are 4 possible moves each move has probability 0.25.

```python
#set up the transition probabilities for states when transition is possible
#set up the policy with equal probability for each possible action
def setupTrasitions():
    random_policy.policy = {}
    for i in range(0,gridHeight):
        for j in range(0,gridWidth):
            states[i][j].transitions = {}
            actions_possible = set()
            if i > 0:
                actions_possible.add(up)
                states[i][j].addTransition(up, states[i-1][j], 1)
            if i < gridHeight-1:
                actions_possible.add(down)
                states[i][j].addTransition(down, states[i+1][j], 1)
            if j > 0:
                actions_possible.add(left)
                states[i][j].addTransition(left, states[i][j-1], 1)
            if j < gridWidth-1:
                actions_possible.add(right)
                states[i][j].addTransition(right, states[i][j+1], 1)
            for action in actions_possible:
                random_policy.updatePolicy(states[i][j], action, 1/len(actions_possible))
setupTrasitions()
```

Having set up the states, actions and the policy of operation, we can now define the environment and the Agent.

```python
#extend the agent to perform policy iterations.
class PolicyIterationAgent(Agent):
    
    def __init__(self, env, policy, theta):
        super().__init__(env, policy)
        self.theta = theta
    
    #policy evaluation step using the bellmans expectation operator
    def _policyEvaluation(self):
        isNotBelowTh = True
        while isNotBelowTh:
            isNotBelowTh = False
            for state in self.env.states:
                oldVal = state.state_value
                newVal = 0
                for action, transitions in state.transitions.items():
                    exp_reward = state.getActionReward(action)
                    discounted_values = 0
                    for next_states, prob in transitions.items():
                        discounted_values += prob[0] * next_states.state_value
                    discounted_values *= self.env.discount
                    newVal += self.policy.getActionProb(state, action) * \
                                (exp_reward + discounted_values)
                state.state_value = newVal
                if abs(oldVal - newVal) > self.theta:
                    isNotBelowTh = True
    
    #policy imporovement step
    def _policyImprovement(self):
        policy_stable = True
        for state in self.env.states:
            bestaction = None
            max_exp_reward = 0
            for action, transitions in state.transitions.items():
                exp_reward = state.getActionReward(action)
                discounted_values = 0
                for next_states, prob in transitions.items():
                    discounted_values += prob[0] * next_states.state_value
                discounted_values *= self.env.discount
                exp_reward += discounted_values
                if exp_reward > max_exp_reward:
                    max_exp_reward = exp_reward
                    bestaction = action
            if bestaction and self.policy.getActionProb(state, bestaction) != 1:
                policy_stable = False
                self.policy.updatePolicy(state, bestaction, 1, reset = True)
        return policy_stable
    
    def policyIteration(self, iterations = True):
        
        if iterations == 0:
            return
        self._policyEvaluation()
        policy_stable = self._policyImprovement()
        while iterations and not policy_stable:
            self._policyEvaluation()
            policy_stable = self._policyImprovement()
            if type(iterations) == int:
                iterations -= 1
                
```

Do policy iteration on the initial random policy to get the optimal strasitions from start to finish.

```python
def policyAgent(discount, iterations = True, showRes = False, reset = True):
    setupTrasitions()
    env = setupEnvironment(discount)
    agent = PolicyIterationAgent(env, random_policy, 0.01)
    agent.policyIteration(iterations)
    reward = simulateAgent(agent, states[0][0], [states[gridHeight-1][gridWidth-1]], showRes = showRes)
    if reset:
        agent.env.reset()
    return reward
```

```python
class ValueIterationAgent(Agent):
    
    def __init__(self, env, policy, epsilon):
        super().__init__(env, policy)
        self.epsilon = epsilon
    #this is the value iteration step i.e. bellman's optimality operateor
    def _update(self):
        maxdiff = 0
        for state in self.env.states:
            max_val = 0
            for action, transitions in state.transitions.items():
                exp_reward = state.getActionReward(action)
                discounted_values = 0
                for next_states, prob in transitions.items():
                    discounted_values += prob[0] * next_states.state_value
                discounted_values *= self.env.discount
                next_state_value = discounted_values + exp_reward
                if next_state_value > max_val:
                     max_val = next_state_value
            if abs(max_val - state.state_value) > maxdiff:
                maxdiff = abs(max_val - state.state_value)
            state.state_value = max_val
        return maxdiff
    
    #update the policy based on the final states of the value update
    def _updatePolicy(self):
        for state in self.env.states:
            bestaction = None
            max_exp_reward = 0
            for action, transitions in state.transitions.items():
                exp_reward = state.getActionReward(action)
                discounted_values = 0
                for next_states, prob in transitions.items():
                    discounted_values += prob[0] * next_states.state_value
                discounted_values *= self.env.discount
                exp_reward += discounted_values
                if exp_reward > max_exp_reward:
                    max_exp_reward = exp_reward
                    bestaction = action
            if bestaction and self.policy.getActionProb(state, bestaction) != 1:
                self.policy.updatePolicy(state, bestaction, 1, reset = True)
    #def update(self):
        
    def valueIteration(self, iterations = True):
        
        if iterations == 0:
            return
        diff = self._update()
        if type(iterations) == int:
            iterations -= 1
        while iterations and diff > self.epsilon * ((1 - self.env.discount)/2*self.env.discount):
            diff = self._update()
            if type(iterations) == int:
                iterations -= 1
        self._updatePolicy()
                
                
```

## 3. Gamblers Problem

#### a. Formulation

<ul>
    <li>States: For each amount possible we assign a state $State(i)$. $State(0)$ defines a failure State. $State(100)$ is the success state. </li>
    <li>Actions: In each state $i$ we can take an action $action(1)$, $action(2)$ ... $action(min(i, 100 - i))$ which corresponds to stakes of the gambler. </li>
    <li>Goals: To reach $State(100)$ starting of with some capital.</li>
    <li>Reward: Reward is $+1$ for $State(100)$ and zero for remaining states. </li>
    <li>Transitions: Given an $action(i)$ and a $State(i)$ we define there is $p$ probability of reaching state $State(i) + action(i)$ (i.e. the gambler wins the bet) and $(1-p)$ probability of reaching state $State(i) - action(i)$ (i.e. loosing the bet).
    <li>Policy; Policy defines which actions to choose in each state. Initialized with a random policy with equal probability on each action.</li>
</ul>

```python
#there are 100 states
states = [State(str(i),0) for i in range(101)]
#there are 50 possible actions overall considering all states.
actions = [Action(str(i)) for i in range(1,51)]
states[100].reward = 1
random_policy = Policy()
```

```python
def defineTransitions(prob):
    for i, state in enumerate(states[1:]):
        state_num = i + 1
        possible_actions = actions[:min(state_num, 100 - state_num)]
        state.transitions = {}
        for a, action in enumerate(possible_actions):
            stake_quatity = a + 1
            state.addTransition(action, states[state_num + stake_quatity], prob)
            state.addTransition(action, states[state_num - stake_quatity], (1 - prob))
            random_policy.updatePolicy(state, action, 1/len(possible_actions))

#defineTransitions(0.3)
```

```python
def defineEnvironment(discount):
    env = Environment(discount)
    for state in states:
        env.addState(state)
    return env

#env = defineEnvironment(1)
```

```python
poissonDict = {}
def poisson(n, lam):
    if (n, lam) not in poissonDict:
        poissonDict[(n, lam)] = ((lam**n)/math.factorial(n))*np.e**(-lam)
    return poissonDict[(n, lam)]
```

```python
states = [[State((j,i),0) for i in range(21)] for j in range(21)]
actions = [Action(i) for i in range(-5,6)]
random_policy = Policy()
fReqMean, sReqMean = 3, 4
fRetMean, sRetMean = 3, 2
```

```python
def getExpectedReward(cars, ReqMean, RetMean, reward = 10, prange = 11):
    if cars == 0:
        return 0
    net_reward = 0
    for i in range(1, prange):
        if i <= cars:
            net_reward += poisson(i, ReqMean) * reward
        else:
            net_reward += poisson(i - cars, RetMean) * poisson(i, ReqMean) * reward
        reward += 10
    return net_reward
    
for i in range(1,21):
    for j in range(1,21):
        states[i][j].reward = getExpectedReward(i, fReqMean, fRetMean) + \
                                getExpectedReward(j, sReqMean, sRetMean)
```

```python
def setup(movecost = 2):
    for i in range(0,21):
        for j in range(0,21):
            action_list = []
            for action in actions:
                a = action.action_name 
                if i - a >= 0 and i - a <= 20 and j + a >= 0 and j + a <= 20:
                    states[i][j].addTransition(action, states[i - a][j + a], 1, 
                                               -1 * movecost * abs(a))
                    action_list += [action]
            for action in action_list:
                random_policy.updatePolicy(states[i][j], action, 1/len(action_list))

setup()
```

```python
def envSetup(discount):
    env = Environment(discount)
    for i in range(21):
        for j in range(21):
            env.addState(states[i][j])
    return env

env = envSetup(0.9)
```
